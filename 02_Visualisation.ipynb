```python
# Téléchargement des données
mnist = fetch_openml('mnist_784', version=1)
X = np.array(mnist['data'])
y = np.array(mnist['target'])
```

```python
# Visualisation en image des données

# Exemple de redimensionnement d'une image
X_reshaped = X.reshape(-1, 28, 28)  # Les images étant des pixels de 28*28

# Affichage de 9 images aléatoires
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_reshaped[i], cmap='gray')  # Affiche l'image en nuances de gris
    ax.set_title(f"Label : {y[i]}")
    ax.axis('off')  

plt.tight_layout()
plt.show()
```
